```python
from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.charts import Bar
from bokeh.models import HoverTool
from bokeh.models.widgets import Select
from bokeh.layouts import column, widgetbox
from bokeh.io import push_notebook,output_notebook, output_file, show
import Tasks2
import pandas as pd
data = Tasks2.final_data
```

```python
output_file("songByYear.html")

p= Bar(data, label='Year', values='Title', agg='count',
        title="Number of Songs by Year", legend=False, plot_width=600, plot_height=400)
show(p)
```

```python
source = data[['Year', 'acousticness', 'Artist', 'Title']]
source.columns = ['x','y','Artist', 'Title']
```

```python
from ipywidgets import interact
output_notebook()
hover = HoverTool(tooltips=[
    ('Title', '@Title'),
    ('Artist', '@Artist'),
    ('(x,y)', '($x,$y)')])

p = figure(title="Year vs Acousticness", tools=[hover])
r = p.circle(x = "x", y = "y",legend = False, source = source)
def update(feature):
    p.title.text = "Year vs " + feature
    r.data_source.data['y'] = data[feature]
    push_notebook()
show(p, notebook_handle=True)
interact(update, feature = ['acousticness','energy','instrumentalness', 'danceability',
                     'key','liveness','loudness','songLength','speechiness','tempo','valence'])
```
